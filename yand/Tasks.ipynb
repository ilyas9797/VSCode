```python
# задание 4
from operator import itemgetter
import sys

Max_Cliques = []
n = 0
m = 0
matrix = []
V = set()


def all_N(u):
    global matrix, n
    neighbors = set()
    for i in range(n):
        if matrix[u - 1][i] == 1:
            neighbors.add(i + 1)
    return neighbors


def choose_max_u(P, X):
    global matrix
    return max([(i, matrix[i - 1]) for i in set.union(P, X)], key=itemgetter(1))[0]

def check_links(v, linked):
    for i in linked:
        if not matrix[v - 1][i - 1]:
            return False
    else:
        return True

def success(R):
    print(' '.join(str(x) for x in R))
    print(' '.join(str(x) for x in V - R))
    sys.exit()
    
def check_clique(R):
    V_without_R = V - R
    if len(V_without_R):
        for v in V_without_R:
            if not check_links(v, set(V_without_R - {v})):
                return
        else:
            success(R)

def dfs(R, P, X):
    global Max_Cliques
    if len(P) == 0 and len(X) == 0:
        Max_Cliques.append(R)
        check_clique(R)
        return
    u = choose_max_u(P, X)
    N_u = all_N(u)
    P_without_N_u = P - N_u
    while len(P_without_N_u) != 0:
        v = P_without_N_u.pop()
        N_v = all_N(v)
        dfs(R | {v}, P & N_v, X & N_v)
        P.discard(v)
        X.add(v)
        P_without_N_u = set.difference(P, N_u)

def task4():
    global n, m, matrix, Max_Cliques, V
    
#     ввод из файла
    with open('4.txt', 'r') as file:
        n, m = map(int, file.readline().split())
        matrix = [[0 for j in range(n)] for i in range(n)]
        for k in range(m):
            i, j = map(int, file.readline().split())
            matrix[i - 1][j - 1] = 1
            matrix[j - 1][i - 1] = 1
    
#   ввод с клавиатуры
#     n, m = map(int, input().split())
#     matrix = [[0 for j in range(n)] for i in range(n)]
#     for k in range(m):
#         i, j = map(int, input().split())
#         matrix[i - 1][j - 1] = 1
#         matrix[j - 1][i - 1] = 1
    
    V = {i + 1 for i in range(n)}
    
    dfs(set(), {i + 1 for i in range(n)}, set())
    print(-1)
```
